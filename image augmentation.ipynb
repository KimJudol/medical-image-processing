```python
import random
import numpy as np
import os
import cv2
import glob
from PIL import Image
import PIL.ImageOps
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
def image_augmented (dir_path):
  filenames = os.listdir(dir_path)
  for filename in filenames:
    file_path = os.path.join(dir_path,filename)
    image = Image.open(file_path)
    aa = filename.split('.')
    aaa = aa[0]

    inverted_image = image.transpose(Image.FLIP_LEFT_RIGHT)
    inverted_image.save(aaa + '_inverted_' + '.png')
    rotated_image = image.rotate(random.randrange(-20, 20),fillcolor=(0,0,0))
    rotated_image.save(aaa + '_rotated_' + '.png')
```

```python
%cd '/content/drive/MyDrive/Classification/original/CN'
image_augmented('/content/drive/MyDrive/Classification/original/CN')
```
